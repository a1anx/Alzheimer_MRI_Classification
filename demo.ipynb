## InceptionResNetV2 Evaluation
We show the loss and accuracy curves for run 1 of InceptionResNetV2. Additionally, we show the confusion matrix.

```python
#Find relative paths
inception_run1_path1 = 'model/inceptionresnetv2/model_output/Run_1/inceptionresnetv2_training_curves.png'
inception_run1_path2 = 'model/inceptionresnetv2/model_output/Run_1/inceptionresnetv2_confusion_matrix.png'

#Load images and set plot structure
img_curves = mpimg.imread(inception_run1_path1)
img_confusion = mpimg.imread(inception_run1_path2)
fig, axes = plt.subplots(2,1, figsize = (12,12))

#Show loss and accuracy curves
axes[0].imshow(img_curves)
axes[0].axis('off')

#Show confusion matrix
axes[1].imshow(img_confusion)
axes[1].axis('off')

#Show plot
plt.show() 
```

Furthermore, we demonstrate the per-class precision, recall, and F1 score of Run 1's model performance on the test dataset.

```python
#Load data and subset relevant tables
df = pd.read_csv('model/inceptionresnetv2/model_output/Run_1/performance_metrics.csv')
incp_resnet_run1_subset = df.iloc[0:4, 0:5]

#Create figure 
fig, ax = plt.subplots(figsize = (8, 5))

#Hide graph axes
ax.axis('off')

#Create table
table = plt.table(cellText = incp_resnet_run1_subset.values, colLabels = incp_resnet_run1_subset.columns, loc = 'center', cellLoc = 'center')

#Display table
plt.show()
```

### Comparative Evaluation of InceptionResNetV2 Training Runs

We can compare the precision, recall, and F1 score of each run to determine the most competitive model.

```python
#Load data and subset relevant tables
df = pd.read_csv('model/inceptionresnetv2/model_output/comparison_metrics.csv')
run_comp_subset = df.iloc[0:5, 0:5]

#Create figure 
fig, ax = plt.subplots(figsize = (8,5))

#Hide graph axes
ax.axis('off')

#Create table
table = plt.table(cellText = run_comp_subset.values, colLabels = run_comp_subset.columns, loc = 'center', cellLoc = 'center')

plt.show()
```

Run 4 achieved the best performance, with test accuracy of 82.4% and test loss of 0.404, significantly outpacing the other models. This indicates that decreasing batch size from 64 to 32 and decreasing learning rate from 0.003 to 0.0001 enabled stable, effective training of the model. Note that run 4 also vastly outperformed other runs in correctly identifying the VeryMildDemented cases, with preciscion score of 0.66, recall of 0.64, and F1-score of 0.65
